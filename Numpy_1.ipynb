Using List
List can not reshape itself but should be converted to numpy array

```python
list_general=[]
list_a=[4,5,6]
list_b=[10,20,30]

# 1D vs 1D list
print(f"Number of elements: {Counter(list_b)}")
print("1D by 1D list")
print(f"""Lists:
list_a: {list_a}
list_b: {list_b}""")
for i,numA in enumerate(list_a):
    for j,numB in enumerate(list_b):
        if i==j:
            list_general.append(numA+numB)
        else:
            continue
print(f"The broadcast Output: {list_general}")
list_d=[
    [1,2,3],
    [4,5,6]
]
print(f"List:              {list_d}")
print(f"Concatenated List: {list_a+list_b}")
# Two Dimensions list with one dimension list
print()
print("2D list vs 1D list")
list_f=[]
for k, element in enumerate(list_d): # break list 
    # print(f"{k}: {element}")
    list_e=[]
    for l,ind_element in enumerate(element):
        # print(ind_element)
        for m,numC in enumerate(list_b):
            if l==m:
                list_e.append(ind_element+numC)
            else:
                continue
    list_f+=list_e
print(f"The Output by +:     {list_f}")
# Reshaping the shape of list
arr=np.array(list_f)
arr=arr.reshape(2,3)
print(f"""Final Output:      
{arr}""")
# print(list_f)
```
