```python
import pandas as pd
import os
from IPython.display import display,Markdown 
import numpy as np
import matplotlib.pylab as plt
```

```python
display(Markdown("### AFRICA POLULATION DATASET CLEANING"))
filepath="Africa_Population.csv"
if os.path.exists(filepath):
    pass
else:
    print(f"File named {filepath} Not Found")
    exit
```

```python
data=pd.read_csv(filepath)
display(Markdown("### DATASET INSPECTION"))
print(f"(Rows,Columns): {data.shape}")
display(Markdown("### FIRST FIVE ROWS"))
display(data.head())
display(Markdown("### LAST FIVE ROWS"))
display(data.tail())
display(Markdown("### DATASET INFO"))
print(data.info())
display(Markdown("### DATASET DESCRIPTION"))
display(data.describe())
```

```python
display(Markdown("## INSPECTINT FOR NULL VALUES AND DUPLICATES"))
display(Markdown("### Inspecting null values"))
if (data.isnull().sum().sum())>0:
    print(data.isnull().sum())
else:
    print("No Null Values ")
if (data.duplicated().sum())>0:
    print(data.duplicated())
else:
    print("No Duplicate values")
```

```python
display(Markdown("## DEALING WITH OUTLIERS"))
display(Markdown("### DETERMINING OUTLIERS"))
AP_P1=data.copy()
for col in data.select_dtypes(["int64","Float64"]).columns:
    Q1=float(data[col].quantile(0.25)) # lower quantile
    Q2=float(data[col].quantile(0.75))  # upper quantile
    IQR=Q2-Q1
    lower_limit=Q1-(1.5*IQR) # lower limit 
    upper_limit=Q2+(1.5*IQR) # upper limit
    print(f"Column Name:{col}")
    print(f"""Lower Limit {lower_limit}
Upper Limit: {upper_limit}""")
    if (data[col]<lower_limit).sum()>0:
        print(f"Lower Outliers:")
        a=data[col]<lower_limit # Return the boolen value
        print(f"{data[col][a]}")
    else:
        print("No Lower Outliers")
    if ((data[col]>upper_limit).sum())>0:
        print(f"Upper Outliers: ")
        a=data[col]>upper_limit # Return the boolen value
        print(f"{data[col][a]}")
    else: 
        print("No Upper Outliers")
    print("")
    AP_P1=AP_P1[(AP_P1[col]>=lower_limit) & (AP_P1[col]<=upper_limit)]
    
```

```python
display(Markdown("### DATA WITHOUT OUTLIERS"))
display(AP_P1.describe())
print("After Outlier removal")
Dime_list=["Rows","Colums"]
a=list(data.shape)
b=list(AP_P1.shape)
for i,dim1 in enumerate(Dime_list):
    for j, num in enumerate(b):
        if i==j:
          print(f"{dim1}:{num}")

print("Original Data")
for i,dim1 in enumerate(Dime_list):
    for j, num in enumerate(a):
        if i==j:
          print(f"{dim1}:{num}")

empty_list=[]
for i,numA in enumerate(a):
    for j, numB in enumerate(b):
        if i==j:
            empty_list.append(numA-numB)
print(f"Dropped rows: {empty_list[0]}")
```

```python
a=np.array(a)
b=np.array(b)
output=a-b
# print(f"Dropeed Rows: {output[0]}")
display(Markdown(f"""
<div style="
color:white;
background-color:blue;
border:5px green solid;
border-radius:15px;
# width:fit-content;
font-size:15px;
">
<p>ALTERNATIVE</p>
<p>Dropeed Rows: {output[0]}</p>
</div>
"""))
```
